### Função round()

* Função round() tem como primeiro parâmetro um número float a ser arredondados. E como segunndo parâmetro quantas casas queremos arrendodas (deve ser informarmano como int)

```python
# Calculadora de IMC
# IMC = peso/(altura²)

altura = float(input("Digite sua altura: "))
peso = float(input("Digite seu peso: "))

imc = peso/(altura ** 2)

print("Seu IMC é: ", round(imc, 2))
```

```python
# Fazer média com 4 notas (float) e exibir a média arredondada dessa notas

nota1 = float(input("Digite a primeira nota: "))
nota2 = float(input("Digite a segunda nota: "))
nota3 = float(input("Digite a terceira nota: "))
nota4 = float(input("Digite a quarta nota: "))

media = (nota1 + nota2 + nota3 + nota4) / 4
media_arredondada = round(media, 2)

print("Sua nota média é:", media_arredondada)

print("\n" )

if media >= 7 :
    print("Usuario aprovado")
elif media >= 5 :
  print("Usuario de DP")
else :
  print("Usuario reprovado")
```

## Print F
* O print(f"") é utilizado para formatação de strings, colocamos a letra "f" colocada com aspas e quando necessesário inserir variáveis ou cálculos matematicosm basta inserirmos dentro de chaves {} os valores/variavez de nosso interesse

* A utilização do :.f, serve para dfinirmos a quantidade de casas decimais

```python
# USANDO PRINT F

nota1 = float(input("Digite a primeira nota: "))
nota2 = float(input("Digite a segunda nota: "))
nota3 = float(input("Digite a terceira nota: "))
nota4 = float(input("Digite a quarta nota: "))

media = (nota1 + nota2 + nota3 + nota4) / 4

print(f"A sua média foi: {media:.2f}")
```

### **Operadores Relacionais:**

- ">" maior: define qual é o maior
- "<" menor: define qual é o menor
- ">=" maior: define qual é o maior ou igual
- "<=" menor: define qual é o menor ou igual

- "==" igual
- "!=" diferente


```python
#Maior
print(5 > 2)
print(2 > 5)

print("A" > "B")
print("Banana" > "Laranja")
```

```python
# Menor
print(5 < 2)
print(29 < 98)
print("Zebra" < "Cachorro")
```

```python
#Maior ou igual
print(10 >= 10)
print(28 >= 24)
print("B" >= "B")
print("Acerola" >= "Jaca")
```

```python
#Menor ou igual
print(980 <= 972)
print(88 <= 88)
print("X" <= "Y")
print("Casa" <= "Carro")
```

```python
#Igual
print("Banana" == "Amarela")
print("Casa" == "casa")
print(1 == -1)
print("G" == "G")
```

```python
#Diferente
print("Pera" != "Pera")
print("Pera" != "pera")
print(3 != 7)
```

```python
# Deem outros exemplos do uso de Operadores Relacionais
print(5 > 2)
print(5 < 2)
print(5 >= 5)
print("Vidro" <= "Vidraça")
print(2 < 3 < 5)
```

### **Operadores Lógicos:**

* and - E (ambas condicionais tem que ser verdadeira para ser true)
* or - OU (uma condicional tem que ser verdadeira para ser true)
* not - NÃO (inverte a condição)

```python
#AND
print(6 > 4 and 9 < 10)
print(5 == 5 and 6 > 10)
print("A" == "B" and "A" < "B")
print("Casa" < "Banana" and "João" == "Maria")
```

```python
#OR
print("A" < "B" or "B" < "A")
print("B" < "A" or "A" < "B")
print(2 < 4 or 5 < 7 or 8 < 20)
```

```python
#NOT
print(not True)

a = "Bola"
print(not a)

print(not False)
```

```python
# Deem outros exemplos do uso de Operadores Lógicos
print(6 > 4 and 9 < 10)
print(6 < 4 or 9 > 10)
print(6 < 4 and 9 == 9)
print(not True)
print(not False)

print(6 < 4 and 9 == 9 or 7 < 6)
if 6 < 4 and 9 == 9:
  print("Verdadeiro")
else:
  print("Falso")
```
