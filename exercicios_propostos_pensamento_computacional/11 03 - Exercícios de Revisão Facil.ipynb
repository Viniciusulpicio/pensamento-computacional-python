#### 1. O Calculador de Idade

Peça para o usuário digitar o ano atual e o ano em que ele nasceu.
Calcule a idade e exiba: "Você tem X anos".

Se a idade for maior ou igual a 18, exiba também: "Você já é maior de
idade".

```python
from datetime import date
ano_atual = date.today().year

ano_nascimento = int(input("Digite seu ano de nascimento: "))
print(f"Sua idade é {ano_atual - ano_nascimento}")

if ano_atual - ano_nascimento >= 18:
  print("Maior de idade")
elif ano_atual - ano_nascimento < 18:
  print("Menor de idade")
else:
  print("Digite uma data valida")
```

#### 6. Calculadora de IMC Simples


O programa deve pedir o peso (kg) e a altura (m) do usuário, calcular o
IMC (Índice de Massa Corporal ) e classificar o resultado conforme a
tabela simplificada:

● Abaixo de 18.5: Abaixo do peso

● 18.5 a 24.9: Peso normal

● 25.0 a 29.9: Sobrepeso

● 30.0 ou mais: Obesidade

Observação: A fórmula para calcular o IMC é a seguinte:
IMC = PESO / (ALTURA)²

```python
peso = float(input("Qual seu peso: "))
altura = float(input("Qual sua altura: "))

imc = peso / (altura**2)
imc_formatado = round(imc, 2)
print(imc_formatado)

if imc <= 18.5:
  print("Abaixo do peso")
elif imc <= 24.9:
  print("Peso normal")
elif imc <= 29.9:
  print("Sobrepeso")
elif imc >= 30:
  print("Obesidade")
else:
  print("error 500")
```

#### 7. Qual Triângulo

Receba três valores que representam os lados de um triângulo.
Primeiro, verifique se esses valores podem formar um triângulo. Se
puderem, diga se o triângulo é:

● Equilátero - 3 lados iguais

● Isósceles - 2 lados iguais

● Escaleno - todos os lados diferentes

Obs.: Para saber se é um triângulo a soma de dois lados deve ser
sempre maior que o terceiro

```python
l1 = float(input("Digite o primeiro lado: "))
l2 = float(input("Digite o segundo lado: "))
l3 = float(input("Digite o terceiro lado: "))

if l1+l2 <= l3 or l1+l3 <= l2 or l2+l3 <= l1:
  print("Não é um triangulo")
else:
  if l1 == l2 and l2 == l3:
    print("Triangulo Equilátero")

  if l1 == l2 or l1 == l3 or l2 == l3:
    print("Triangulo Isósceles")

  if l1 != l2 or l1 != l3 or l2 != l3:
    print("Triangulo Escaleno")
```

#### 11.​ Validador de Ano Bissexto

Peça um ano. O programa deve dizer se ele é bissexto.

Regra: Um ano é bissexto se for divisível por 4, mas não por 100, a
menos que também seja divisível por 400.

```python
ano = int(input("Digite o ano: "))

if ano % 4 == 0 and ano % 100 != 0:
    print("Ano bissexto")
elif ano % 400 == 0:
    print("Ano bissexto")
else:
    print("O ano não é bissexto")
```
