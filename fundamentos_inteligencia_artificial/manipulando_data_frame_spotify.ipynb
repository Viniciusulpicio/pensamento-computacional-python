```python
df['artist_name']

contagem = df['artist_name'].value_counts()
print(contagem)
```

```python
contagem_df = df['artist_name'].value_counts().reset_index()

contagem_df.columns = ['artist', 'frequencia']

contagem_df
```

```python
contagem_df.plot()
```

```python
print("\n printando por tipo de barra")
contagem_df.head().plot(kind='bar')

print("\n printando por tipo de linha")
contagem_df.head().plot(kind='line')

print("\n printando por tipo de ")
contagem_df.head().plot(kind='hist')

print("\n printando por tipo de caixa")
contagem_df.head().plot(kind='box')

print("\n printando por tipo de frequencia")
contagem_df.head().plot(kind='kde')

print("\n printando por tipo de area")
contagem_df.head().plot(kind='area')
```
