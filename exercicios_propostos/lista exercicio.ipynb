### Exercício 02 - Média de duas notas

Crie um algoritmo que peça ao usuário o nome do aluno, as notas da P1 e da P2 e calcule a
média aritmética.


O algoritmo deve continuar executando esse processo para vários alunos, até que o usuário
decida finalizar.


Ao final de cada cálculo, pergunte ao usuário se deseja continuar ou encerrar, se ele
escolher encerrar, imprima uma lista no formato:

nome_do_aluno - media_final

```python
lista = []

while True:
    nome = str(input("Digite o nome do aluno: "))
    p1 = float(input("Digite a nota da sua P1: "))
    p2 = float(input("Digite a nota da sua P2: "))
    lista.append({'nome': nome, 'media': (p1+p2)/2})


    loop = str(input("\nDeseja continuar ou encerrar o programa (encerrar | continuar)")).upper().strip()
    if loop == "ENCERRAR":
      for aluno in lista:
        print(f"{aluno['nome']} - {aluno['media']:.2f}")
      break
```

### Exercício 04 - Cálculo da Hipotenusa

Crie um algoritmo que receba do usuário o valor dos dois catetos (lados menores do
triângulo) e calcule a hipotenusa de um triângulo retângulo.

A fórmula para o cálculo da hipotenusa é: hipotenusa = √(cateto1² + cateto2²)


```python
import math

cateto_oposto = float(input("Digite o valor do cateto oposto: "))
cateto_adjacente = float(input("Digite o valor do cateto adjacente: "))

print(f"O valor da sua hipotenusa é: {math.sqrt(cateto_oposto**2 + cateto_adjacente**2)}")
```

### Exercício 05 - Salário com aumento

Crie um algoritmo que receba do usuário o valor do salário de um funcionário e calcule o
novo salário com aumento de 3.2%.

O algoritmo deve continuar executando esse processo para vários funcionários, até que o
usuário decida finalizar.


```python
while True:
  salario = float(input("Digite o valor do seu salário: "))
  aumento = salario*0.032
  print(f"Seu salário bonificado é {salario+aumento}")

  sair = str(input("\nCaso queria sair digite (sair|continuar)")).upper().strip()
  if sair == "SAIR":
    break
```

### Exercício 08 - Maior de dois números

Crie um algoritmo que receba do usuário dois números inteiros e mostre qual deles é o
maior. Caso sejam iguais, informe que são iguais.

O algoritmo deve continuar executando esse processo várias vezes, até que o usuário
decida finalizar.


```python
while True:
  num1 = int(input("Digite o primeiro numero: "))
  num2 = int(input("Digite o segundo numero: "))

  if num1 == num2:
    print("Os numeros são iguais")
  elif num1 > num2:
    print(f"O primeiro numero {num1} é maior que {num2}")
  else:
    print(f"O segundo numero {num2} é maior que {num1}")

  sair = str(input("\nDigite `sair` caso queira encerrar o processo ou `continar` (sair|continuar)")).upper().strip()

  if sair == "SAIR":
      break
```

### Exercício 09 - Triagem de temperatura

Crie um algoritmo que receba do usuário a temperatura corporal de uma pessoa e informe:

● menor que 35: Temperatura muito baixa. Pulseira vermelha.

● entre 35 e 37.5: Temperatura normal. Pulseira branca.

● acima de 37.5: Febre. Pulseira azul.


```python
temperatura = float(input("Digite sua temperatura: "))

if temperatura < 35:
  print("Temperatura muito baixa. Pulseira vermelha.")
elif temperatura > 37.5:
  print("Febre. Pulseira azul.")
elif temperatura > 35:
  print("Temperatura normal. Pulseira branca.")
else:
  print("Sem ação programada para essa temperatura")
```

### Exercício 10 - Promoção por valor de compra

Crie um algoritmo que receba do usuário o valor total de uma compra:

● se for acima de R$ 100, aplique 10% de desconto

● caso contrário, aplique 5% de desconto

Mostre o valor final a ser pago pelo usuáriO.

```python
total = float(input("Digite o valor total da sua compra: "))

if total >= 100:
  print(f"Voce ganhou um desconto de 10% agora sua compra ficou {total-total*0.1}")
else:
  print(f"Voce ganhou um desconto de 5% agora sua compra ficou {total-total*0.05}")
```

### Exercício 11 - Soma dos números de 1 até N

Leia um número inteiro N e calcule a soma de todos os números de 1 até N.

Exemplo:
Se N = 5, então: 1 + 2 + 3 + 4 + 5 = 15

```python
#formula de Gauss
num = int(input("Digite um número: "))
soma = num * (num + 1) // 2
print(f"A soma de 1 até {num} é: {soma}")
```

### Exercício 12 - Tabuada

Crie um algoritmo que receba do usuário um número e mostre a tabuada dele de 1 até 10.

Exemplo:
7 x 1 = 7
7 x 2 = 14
…

O algoritmo deve continuar executando esse processo até que o usuário informe o número
-1.


```python
while True:
  tabuada = int(input("Digite o numero que quer verificar a tabuada: "))

  for i in range (11):
    print(f"{tabuada} x {i} = {tabuada*i}")

  parar = str(input("\nCaso queira encerrar o programa digite `-1` ou continunar (-1|continuar)")).strip()
  if parar == "-1":
    break
```
