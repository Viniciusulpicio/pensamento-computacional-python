# Projeto de Análise de Dados Público

## Etapa 1: Escolha do conjunto de dados
O conjunto de dados escolhido é o **Titanic Dataset**, obtido via biblioteca Seaborn.
- **Mínimo de 1000 linhas:** Utilizaremos o dataset completo para análise.
- **Variáveis relevantes:** surivived, pclass, sex, age, sibsp, parch, fare, embarked, etc.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Carregando o dataset
df = sns.load_dataset('titanic')

## Etapa 2: Exploração inicial (EDA)
print("--- Primeiras 5 linhas ---")
print(df.head())

print("\n--- Informações do Dataset ---")
print(df.info())

print("\n--- Estatísticas Descritivas ---")
print(df.describe())
```

### Limpeza de Dados
Identificando valores ausentes e realizando limpeza básica.

```python
# Verificando nulos
print("Nulos por coluna:")
print(df.isnull().sum())

# Limpeza básica
df['age'] = df['age'].fillna(df['age'].median())
df.drop(columns=['deck'], inplace=True, errors='ignore')
df.dropna(subset=['embarked'], inplace=True)

print("\nLimpeza concluída.")
```

## Etapa 3: Visualização de dados
Criando gráficos para análise visual.

```python
plt.figure(figsize=(15, 10))

# 1. Histograma de Idade
plt.subplot(2, 3, 1)
sns.histplot(df['age'], kde=True, color='skyblue')
plt.title('Distribuição de Idade')

# 2. Sobrevivência por Sexo
plt.subplot(2, 3, 2)
sns.barplot(x='sex', y='survived', data=df)
plt.title('Taxa de Sobrevivência por Sexo')

# 3. Tarifa por Classe
plt.subplot(2, 3, 3)
sns.boxplot(x='pclass', y='fare', data=df)
plt.title('Tarifa por Classe')

# 4. Idade vs Tarifa
plt.subplot(2, 3, 4)
sns.scatterplot(x='age', y='fare', hue='survived', data=df)
plt.title('Idade vs Tarifa')

# 5. Heatmap
plt.subplot(2, 3, 5)
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='RdBu')
plt.title('Correlação')

plt.tight_layout()
plt.show()
```

## Etapa 4: Análise

1. **Tendências:** Mulheres tiveram uma taxa de sobrevivência significativamente maior.
2. **Influência:** A classe econômica (pclass) influenciou diretamente a chance de sobrevivência e o valor da tarifa.
3. **Outliers:** Existem tarifas extremamente altas na 1ª classe que fogem da média do dataset.

## Etapa 5: Conclusões
Aprendemos que fatores socioeconômicos e gênero foram determinantes no desastre do Titanic. A análise demonstra como dados estruturados podem revelar padrões históricos importantes.
