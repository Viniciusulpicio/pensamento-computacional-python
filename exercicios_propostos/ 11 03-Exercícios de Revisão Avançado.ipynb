### 1. Validação de senha forte

Leia uma senha e verifique se ela atende a todas as regras:

● mínimo de 8 caracteres

● pelo menos 1 letra maiúscula

● pelo menos 1 letra minúscula

● pelo menos 1 número

● pelo menos 1 caractere especial

Mostre quais critérios foram atendidos e quais não foram.

```python
print("Sua senha deve ter no minimo 8 caracteres, 1 letra maiúscula, 1 letra minúscula, 1 numero e 1 caracter especial \n")
password = str(input("Digite sua senha: "))

if len(password) < 8:
  print("Senha tem menos que 8 caracteres")
elif password == password.lower():
  print("Precisa de uma letra maiúscula")
elif password == password.upper():
  print("Precisa de uma letra minúscula")
elif any(char.isdigit() for char in password) == False:
  print("Precisa de um número")
elif password.isalnum():
  print("Precisa de um caracter especial")
else:
  print("Passou por todos requisitos")

```

### 2. Caixa eletrônico com restrições

Leia um valor inteiro para saque.
O algoritmo deve informar a quantidade mínima de notas para
representar esse valor e garantir que o valor do saque será possível de
ser feito, considerando notas de:

● 100

● 50

● 20

● 10

● 5

● 2

Se o valor for menor ou igual a zero, rejeitar.
Mostrar o total de cédulas usadas para o saque.


```python
saque = int(input("Digite o valor do saque: "))

if saque <= 0 or saque == 1 or saque == 3:
    print("Valor inválido ou impossível de sacar.")
else:
    n100, resto = divmod(saque, 100)
    n50, resto = divmod(resto, 50)
    n20, resto = divmod(resto, 20)
    n10, resto = divmod(resto, 10)
    n5, resto = divmod(resto, 5)
    n2, resto = divmod(resto, 2)

    total_notas = n100 + n50 + n20 + n10 + n5 + n2

    print(f"Notas de 100: {n100}")
    print(f"Notas de 50:  {n50}")
    print(f"Notas de 20:  {n20}")
    print(f"Notas de 10:  {n10}")
    print(f"Notas de 5:   {n5}")
    print(f"Notas de 2:   {n2}")
    print(f"Total de cédulas: {total_notas}")
```

#### 3. Controle de acesso

Leia a idade e se a pessoa possui documento (S ou N).

O sistema deve informar:

● Entrada permitida, se tiver 18 anos ou mais e possuir documento

● Entrada negada por idade se tiver menos que 18 anos

● Entrada negada por falta de documento se estiver sem documento

● Entrada negada por ambos

```python
idade = int(input("Digite sua idade: "))
documento = str(input("Digite se tem documento (S ou N): "))
documento_limpo = documento.upper().strip()

if idade >= 18 and documento_limpo == "S" and documento != "SIM":
  print("Entrada permitida")
elif idade < 18:
  print("Entrada negada por idade")
elif documento_limpo != "S" or "SIM":
  print("Entrada negada por falta de documento")
elif documento != "S" and documento != "SIM" and idade < 18:
  print("Entrada negada por falta de documento e idade")
else:
  print("505: erro")
```

#### 4. Forma de pagamento de compra

Leia o valor de uma compra e a forma de pagamento:

● 1 = à vista em dinheiro

● 2 = à vista no cartão

● 3 = parcelado em 2x

● 4 = parcelado em 3x ou mais

● dinheiro: 10% de desconto

● cartão à vista: 5% de desconto

● 2x: preço normal

● 3x ou mais: 20% de juros

● se for 3x ou mais, pedir o número de parcelas.

● valor final

● valor de cada parcela, quando houver


```python
valor_compra = float(input("Digite o valor da compra: "))
forma_pagamento = int(input("Digite a forma de pagamento: \n 1 = à vista em dinheiro \n 2 = à vista no cartão \n 3 = parcelado em 2x \n 4 = parcelado em 3x ou mais \n"))

if forma_pagamento == 1:
  desconto = valor_compra*0.10
  preco_final = valor_compra - desconto
  print(f"O valor da compra ficou: {preco_final}")
elif forma_pagamento == 2:
  desconto = valor_compra*0.05
  preco_final = valor_compra - desconto
  print(f"O valor da compra ficou: {preco_final}")
elif forma_pagamento == 3:
  valor_compra = valor_compra
  print(f"O valor da compra ficou: {valor_compra}")
elif forma_pagamento == 4:
  n_parcelas = int(input("Digite o número de parcelas: "))
  print(f"\n O valor da compra ficou: {valor_compra} e em {n_parcelas} parcelas")
else:
  print("505: error")
```
