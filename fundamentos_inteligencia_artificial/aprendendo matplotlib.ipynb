```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('/content/sample_data/california_housing_train.csv')
df
```

```python
df['median_house_value'].hist(bins = 50)
plt.title('Distribuição de valores')
plt.xlabel('Valor')
plt.ylabel('Frequencia')

# no computador ele não mostraria a imagem tem que dar show
plt.show()
```

```python
df.plot.scatter(x = 'median_income', y='median_house_value')
plt.title('Distribuição de valores')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.show()
```

```python
df.plot.scatter(x = 'longitude', y='latitude')
plt.title('Distribuição de valores')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.show()
```

```python
df.plot.scatter(x = 'median_income', y='median_house_value', c='housing_median_age', cmap='plasma')
plt.title('Exemplo com scartter')
plt.xlabel('Valor médio casa')
plt.ylabel('Renda média familiar')
plt.show()
```

```python
avg_house_value_by_age = df.groupby('housing_median_age')['median_house_value'].mean()
avg_house_value_by_age.plot(kind = 'bar')

plt.title('Preço médio por ano')
plt.xlabel('Ano')
plt.ylabel('Preço')
plt.show()
```

```python
plt.figure(figsize=(8,8))
scatter = plt.scatter(x=df['longitude'], y=df['latitude'], c=df['median_house_value'])

plt.colorbar(scatter, label = "Median house value")
plt.title('Pocicoes')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
corraletion_matrix = df.corr()
```

```python
plt.figure(figsize=(12, 8 ))
plt.matshow(corraletion_matrix, cmap='coolwarm')

plt.xticks(range(len(corraletion_matrix.columns)), list(corraletion_matrix.columns))
```
