# Objetivo 3

- saber quais partidas duram mais em base o número de rounds que acontecem por mapa


```python
df['total_rounds'] = df['result_1'] + df['result_2']

plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='_map', y='total_rounds')

plt.title('Duração das Partidas por Mapa (Total de Rounds Jogados)', fontsize=16)
plt.xlabel('Nome do Mapa')
plt.ylabel('Quantidade Total de Rounds')
plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.histplot(data=df, x='total_rounds', bins=40, kde=True)

plt.title('Distribuição da Duração das Partidas (Histograma)')
plt.xlabel('Total de Rounds na Partida')
plt.ylabel('Frequência (Quantidade de Partidas)')

plt.xlim(15, 60)

plt.show()
```

- nessa análise o que acontece é a soma dos 2 halfs da partida de CS para saber o total de round daquela partida, depois usa o seaborn para fazer um Boxplot com o df criado. Esse Boxplot funciona assim, a linha do meio da caixa é a mediana e toda a caixa azul é onde esta contentrado a maior proporção dos dados, a linha de baixo é o menor e o de cima maior. O diferencial são as "bolinhas" depois das linhas limitando, elas mostram os diferenciis conhecidos como Outliers ou Valores Discrepantes.

### Analise

- Conseguimos tirar do gráfico que na Mirage chegaram partidas a quase 80 rounds podendo ser um dos mapas mais equilibrados para se jogar tendo em vista que no OT fica trocando de lado toda hora, além de que na train foi o mapa com menos round podendo indicar que menos times dominam o mapa e quem conhece tem vantagem. Outra análise é que no mapa da Vertigo é o que tem menos variação podendo indicar que o jogo acaba geralmente no tempo regulamentar e com os times equilibrados

# Objetivo 4

- times com maior numeros de vitórias



```python
# não sei muito bem usar essa lib numpy mas consigo usar vetorização pegando todos os dados da coluna match_winne
# e fazendo a logica do where na coluna inteira e não linha por linha
df['nome_vencedor_partida'] = np.where(
    df['match_winner'] == 1,
    df['team_1'],
    df['team_2']
)

maiores_campeoes = df['nome_vencedor_partida'].value_counts().head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=maiores_campeoes.index, y=maiores_campeoes.values)

plt.title('Top 10 Maiores Campeões (Vitórias em Partidas)')
plt.xlabel('Nome da Equipe')
plt.ylabel('Quantidade de Vitórias')
plt.xticks(rotation=45)

plt.show()
```

- aqui usamos a biblioteca numpy para comparar a coluna toda e desfazer o mapeamento condicional e voltar para os nomes nas colunas e depois usa o seaborn para plotar em barra

### Analise

- Conseguimos tirar do gráfico que comprova nossa tese do objetivo 2 que a Liquid e Astralis sendo o time que mais se enfrentaram também são os que tem mais vitória podendo indicar que ambas sempre avançaram muito nas competiçoes e consequentemente jogando muitas partidas e ganhando muitas

# Objetivo 5

- aqui queremos ánalisar quais as vantagens do lado ct/tr em cada mapa


```python
df['total_ct'] = df['ct_1'] + df['ct_2']
df['total_tr'] = df['t_1'] + df['t_2']

tabela_lados = df.groupby('_map')[['total_ct', 'total_tr']].mean()

plt.figure(figsize=(10, 8))

sns.heatmap(data=tabela_lados, annot=True, fmt='.2f')

plt.title('Vantagem de Lado: Média de Rodadas (CT vs TR)')
plt.xlabel('Lado')
plt.ylabel('Nome do Mapa')

plt.show()
```

- aqui somamos os pontos de ct e tr de ambos os times para minimizar o viés de habilidade de time, e somamos todos os pontos de cada mapa e tiramos a média e com o seaborn monta a tabela de calor com mas médias de vitoria em cada lado de cada mapa

### Analise

- Conseguimos tirar do gráfico que os mapas menos equilibrados são Train e Nuke tendo uma desvantagem muito grande de tr e uma vantagem grande de ct e o mapa da inferno sendo o mais equilibrado entre os lados. Apesar de conseguirmos análisar esse grafico ele pode ter viés como quem começa em determinado lado pode ter uma vantagem e assim aumentar sua economia e consequentemente ganhar mais rounds do mesmo lado desestabilizando seu oponente mentalmente porém nos somamos os rounds de ambos os times para que esse viés diminua um pouco tirando o lado da habilidade de cada time.
