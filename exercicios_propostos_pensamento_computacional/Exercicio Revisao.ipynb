### Exercício 6

Crie duas variáveis numéricas e:

● Some

● Subtraia

● Multiplique

● Divida

Imprima todos os resultados.

```python
num1 = float(input("Digite o primeiro numero: "))
num2 = float(input("Digite o segundo numero: "))

soma = (num1+num2)
sub = (num1-num2)
mult = (num1*num2)
div = (num1/num2)

print("A soma dos numeros é", soma)
print("A subtração dos numeros é", sub)
print("A multiplicação dos numeros é", mult)
print("A divisão dos numeros é", div)
```

### Exercício 8

Peça ao usuário a largura e altura de um retângulo.
Calcule e exiba a área.

Área de um retângulo = Base x Altura

```python
base = float(input("Digite a base do retângulo: "))
altura = float(input("Digite a altura do retângulo: "))

area = base*altura
print("A área do retângulo é:", area)
```
