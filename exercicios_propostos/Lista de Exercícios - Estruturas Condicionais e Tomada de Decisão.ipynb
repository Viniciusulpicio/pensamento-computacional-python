## Exercício 1:

Crie um fluxograma e escreva um algoritmo para um caixa eletrônico de
um banco, que permita que o usuário realize um saque no valor que ele
informar.


Requisitos:
- Perguntar ao cliente qual o valor que ele deseja sacar?
- Verificar se esse cliente possui saldo
- Se o cliente possuir saldo, permitir o saque
- Senão, informá-lo que o saldo é insuficiente

```python
import random

saldo = random.randint(1, 1000000)

print("seu saldo é: ", saldo)
saque = float(input("Digite o valor do saque desejado: "))

if saque < saldo:
  saldo = saldo - saque
  print("Saque de:", saque, "realizado com sucesso! Agora seu saldo é: ", saldo)
elif saque > saldo:
  print("Sua tentativa de saque é maior que seu saldo. Seu saldo é de: ", saldo)
else:
  print("error 500")
```

### Exercício 4:

Escreva um algoritmo que ajude o médico a diagnosticar seu paciente
com base na triagem que foi feita quando ele chegou no hospital.

As condições são:

● Tem Febre, dor no corpo e o nariz está escorrendo = Gripe

● Tem Febre, dor no corpo e diarreia = Virose

● Tem Febre, dor no corpo e não está sentindo cheiro = Covid

● Está com dor de cabeça muito forte, mas não tem outros sintomas

```python
print("Digite 'sim' ou 'nao' para as seguintes perguntas:")
print("\n")

febre = input("O paciente esta com febre? ").lower().strip()
dor_corpo = input("O paciente esta com dor no corpo? ").lower().strip()
nariz_congestionado = input("O paciente esta com nariz congestionado? ").lower().strip()
sem_sentir_cheiro = input("O paciente esta sem sentir cheiro? ").lower().strip()
dor_cabeça = input("O paciente esta com dor de cabeça? ").lower().strip()
diarreia = input("O paciente esta com diarreia? ").lower().strip()

if febre == "sim" and dor_corpo == "sim" and nariz_congestionado == "sim":
  print("Usuario com gripe")
elif febre == "sim" and dor_corpo == "sim" and diarreia == "sim":
  print("Usuario está com virose")
elif febre == "sim" and dor_corpo == "sim" and sem_sentir_cheiro == "sim":
  print("Usuario está com covic")
elif dor_cabeça == "sim":
  print("Usuario esta com enxaqueca")
else:
  print("Não conseguimos diagnosticar o paciente")
```

### Exercício 5:
Crie um programa que peça um número inteiro ao usuário e diga se ele
é par ou ímpar.

Dica: Use o operador de módulo % (resto da divisão). Se o resto da
divisão por 2 for zero, o número é par.

```python
numero = int(input("Digite um número para descobrir se é par ou impar: "))

if numero%2 == 0:
  print("O número é par")
else:
  print("O numero é impar")
```

### Exercício 7:

O programa deve pedir o peso (kg) e a altura (m) do usuário, calcular o
IMC (Índice de Massa Corporal ) e classificar o resultado conforme a
tabela simplificada:

● Abaixo de 18.5: Abaixo do peso

● 18.5 a 24.9: Peso normal

● 25.0 a 29.9: Sobrepeso

● 30.0 ou mais: Obesidade

Observação: A fórmula para calcular o IMC é a seguinte:
IMC = PESO / (ALTURA)²

```python
peso = float(input("Qual seu peso: "))
altura = float(input("Qual sua altura: "))

imc = peso / (altura**2)
imc_formatado = round(imc, 2)
print(imc_formatado)

if imc <= 18.5:
  print("Abaixo do peso")
elif imc <= 24.9:
  print("Peso normal")
elif imc <= 29.9:
  print("Sobrepeso")
elif imc >= 30:
  print("Obesidade")
else:
  print("error 500")
```

### Exercício 8:

Um professor precisa de um sistema que receba 3 notas de um aluno. O
programa deve calcular a média e exibir:

● Média >= 7.0: "Aprovado! Parabéns."

● Média entre 5.0 e 6.9: "Em recuperação."

● Média abaixo de 5.0: "Reprovado."

```python
nota1 = float(input("Digite a primeira nota: "))
nota2 = float(input("Digite a segunda nota: "))
nota3 = float(input("Digite a terceira nota: "))

media = (nota1+nota2+nota3)/3

if media >= 7.0:
  print("Aprovado! Parabéns.")
elif media < 6.9 and media >= 5:
  print("Em recuperação.")
elif media < 5:
  print("Reprovado")
else:
  print("error: 500")
```

### Exercício 10 - Desafio

Você foi contratado para programar a lógica de um elevador de carga
industrial. O sistema deve garantir que o elevador não opere em
condições perigosas e que apenas pessoal autorizado tenha acesso a
certas funções.

### Requisitos do Sistema:

O programa deve solicitar ao usuário as seguintes informações:

1. Peso da Carga (em kg).

2. Peso da Cabine (em kg).

3. Status da Porta (0 para aberta, 1 para fechada).

4. Senha de Operador (um número inteiro).

#### Regras:

O elevador só deve exibir a mensagem "SISTEMA OPERACIONAL" se
todas as condições abaixo forem verdadeiras:

● Segurança de Peso: O Peso Total (Carga + Cabine) não pode
exceder 2000 kg.

● Segurança Física: A porta deve estar obrigatoriamente fechada.

● Autenticação: A senha inserida deve ser igual a 1234.

Caso contrário, o programa deve exibir "SISTEMA BLOQUEADO" e
indicar o motivo (ex: "Excesso de peso" ou "Porta aberta").

#### Entrega:

Faça um fluxograma e implemente um algoritimo que atenda os requisitos e regras descritos acima.

```python
carga = float(input("Digite o peso da carga em kg: "))
cabine = float(input("Digite o peso da cabine em kg: "))
status_porta = int(input("Digite o status da porta: 0 = aberta e 1 = fechada: 1"))
senha = int(input("Digite a senha do operador: "))

peso_total = cabine+carga

if peso_total <= 2000 and status_porta == 1 and senha == 1234:
  print("SISTEMA OPERACIONAL")
else:
  if peso_total > 2000:
    print("SISTEMA BLOQUEADO: Excesso de peso")
  elif status_porta == 0:
    print("SISTEMA BLOQUEADO: Porta aberta")
  elif senha != 1234:
    print("SISTEMA BLOQUEADO: Senha incorreta")
  else:
    print("error: 500")
```
