### Exercício 02 - Média de duas notas

Crie um algoritmo que peça ao usuário o nome do aluno, as notas da P1 e da P2 e calcule a
média aritmética.


O algoritmo deve continuar executando esse processo para vários alunos, até que o usuário
decida finalizar.


Ao final de cada cálculo, pergunte ao usuário se deseja continuar ou encerrar, se ele
escolher encerrar, imprima uma lista no formato:

nome_do_aluno - media_final

```python
lista = []

while True:
    nome = str(input("Digite o nome do aluno: "))
    p1 = float(input("Digite a nota da sua P1: "))
    p2 = float(input("Digite a nota da sua P2: "))
    lista.append({'nome': nome, 'media': (p1+p2)/2})


    loop = str(input("\nDeseja continuar ou encerrar o programa (encerrar | continuar)")).upper().strip()
    if loop == "ENCERRAR":
      for aluno in lista:
        print(f"{aluno['nome']} - {aluno['media']:.2f}")
      break
```

### Exercício 19 - Cadastro de alunos com análise final

Crie um algoritmo que receba do usuário o nome e a nota de 5 alunos.

Ao final, mostre:

● o nome do aluno com a maior nota

● o nome do aluno com a menor nota

● a média da turma

● quantos alunos foram aprovados

Considere aprovado quem tiver nota maior ou igual a 7.


```python
nomes = []
notas = []
aprovados = 0
soma_notas = 0

for i in range(5):
  nome = str(input(f"\nDigite o nome do {i+1}º aluno: "))
  nota = float(input(f"Digite a nota do {i+1}º aluno: "))

  nomes.append(nome)
  notas.append(nota)

  soma_notas += nota
  if nota >= 7:
    aprovados = aprovados+1

maior_nota = max(notas)
menor_nota = min(notas)
media = soma_notas/5

indice_maior = notas.index(maior_nota)
indice_menor = notas.index(menor_nota)

print(f"O aluno com maior nota é: {nomes[indice_maior]} {maior_nota}")
print(f"O aluno com menor nota é: {nomes[indice_menor]} {menor_nota}")
print(f"A média da turma foi: {media}")
print(f"O numero de alunos aprovados foi: {aprovados}")
```

Crie um algoritmo que simule um pequeno sistema de lanchonete.

● 3 = Refrigerante = R$ 7

● 4 = Milk-shake = R$ 15

● 5 = Encerrar pedido

O usuário pode fazer vários pedidos até escolher encerrar. No final, mostre:

● valor total do pedido

● valor do couvert - 5%

● valor da taxa do garçom - 10%

● valor total a pagar (valor total do pedido + valor do couvert + valor da taxa do garçom)

Pare o algoritmo quando a porcentagem de bateria chegar a 0 ou menos e informe quantos usos foram feitos.

### Exercício 02 - Analisador de texto simples

Crie um algoritmo que analise textos. Peça para que o usuário digite uma frase, seu algoritmo deve:

● contar quantas letras a frase tem

● contar quantos espaços a frase tem

● contar quantos números aparecem na frase

E retornar o resultado na tela para o usuário.


```python
frase = str(input("Digite sua frase: "))
espacos = 0
letras = 0
numeros = 0

for char in frase:
  if char.isspace():
    espacos += 1
  elif char.isalpha():
    letras += 1
  elif char.isdigit():
    numeros += 1

print(f"Espaços: {espacos}, Letras: {letras}, Números: {numeros}")
```

### Exercício 03 - Jogo de adivinhação com limite

Crie um algoritmo para um jogo de adivinhação com limite de tentativas. Seu algoritmo deve seguir o seguinte passo a passo:

    ● Gerar um número inteiro fixo e aleatório entre 1 e 500

      ○ Para gerar o número aleatório:

        ■ importe a biblioteca random - import random

        ■ use a função randint da biblioteca - random.randint(1, 500)

        ● Peça para que o usuário tente adivinhar o número

    ● Retorne ao usuário se o número é maior, menor ou se ele acertou

    ● O limite de tentativas deve ser 5.


```python
import random
aleatorio = random.randint(1, 500)

for i in range (5):
  tentativa = int(input("Tente adivinhar o numero (entre 1 e 500): "))

  if tentativa == aleatorio:
    print("Você acertou !")
    break
  elif tentativa > aleatorio:
    print("O numero é menor que sua tentativa")
  elif tentativa < aleatorio:
    print("O numero é maior que sua tentativa")
  else:
    print("Error")

if tentativa != aleatorio:
  print(f"\nO numero correto era {aleatorio}")
```
