### 1. Validação de senha forte

Leia uma senha e verifique se ela atende a todas as regras:

● mínimo de 8 caracteres

● pelo menos 1 letra maiúscula

● pelo menos 1 letra minúscula

● pelo menos 1 número

● pelo menos 1 caractere especial

Mostre quais critérios foram atendidos e quais não foram.

```python
print("Sua senha deve ter no minimo 8 caracteres, 1 letra maiúscula, 1 letra minúscula, 1 numero e 1 caracter especial \n")
password = str(input("Digite sua senha: "))

if len(password) < 8:
  print("Senha tem menos que 8 caracteres")
elif password == password.lower():
  print("Precisa de uma letra maiúscula")
elif password == password.upper():
  print("Precisa de uma letra minúscula")
elif any(char.isdigit() for char in password) == False:
  print("Precisa de um número")
elif password.isalnum():
  print("Precisa de um caracter especial")
else:
  print("Passou por todos requisitos")

```

#### 7. Detector de palíndromo numérico

Leia um número inteiro positivo e verifique se ele é um palíndromo
(número que se lê da mesma forma da esquerda para a direita e da
direita para a esquerda, desconsiderando espaços e pontuação).

Exemplos:

● 121 → palíndromo

● 1331 → palíndromo

● 123 → não é palíndromo

Obs.: Fazer sem converter para texto.

```python
numero = int(input("Digite um número qualquer: "))
numero_original = numero
invertido = 0

while numero > 0:
    digito = numero % 10
    invertido = (invertido * 10) + digito
    numero = numero // 10

if invertido == numero_original:
    print("É um palíndromo")
else:
    print("Não é um palíndromo")
```

#### 10. Qual Triângulo

Receba três valores que representam os lados de um triângulo.
Primeiro, verifique se esses valores podem formar um triângulo. Se
puderem, diga se o triângulo é:

● Equilátero - 3 lados iguais

● Isósceles - 2 lados iguais

● Escaleno - todos os lados diferentes

Obs.: Para saber se é um triângulo a soma de dois lados deve ser
sempre maior que o terceiro

```python
l1 = float(input("Digite o primeiro lado: "))
l2 = float(input("Digite o segundo lado: "))
l3 = float(input("Digite o terceiro lado: "))

if l1+l2 <= l3 or l1+l3 <= l2 or l2+l3 <= l1:
  print("Não é um triangulo")
else:
  if l1 == l2 and l2 == l3:
    print("Triangulo Equilátero")

  if l1 == l2 or l1 == l3 or l2 == l3:
    print("Triangulo Isósceles")

  if l1 != l2 or l1 != l3 or l2 != l3:
    print("Triangulo Escaleno")
```
