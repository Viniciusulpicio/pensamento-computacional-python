```python
requisitos_obrigatorios = {"python", "sql", "git", "aws", "js", "node"}

print("Analisador de Requisitos para Vaga Backend")
print(f"Requisitos da vaga: {requisitos_obrigatorios}\n")

entrada = input("Digite suas habilidades técnicas separadas por vírgula : ")

# função set é a funcao do python para se trabalhar com conjuntos (mesma do js)
habilidades_candidato = set(item.strip().lower() for item in entrada.split(","))

interseção = requisitos_obrigatorios.intersection(habilidades_candidato)

# vagas / candidato
diferenca = requisitos_obrigatorios.difference(habilidades_candidato)

# candidato / vagas
diferencial = habilidades_candidato.difference(requisitos_obrigatorios)

print("\n ESULTADO DA ANÁLISE ")
print(f" Interseção : {interseção if interseção else '∅'}")
print(f" Diferença Vaga - Candidato (O que falta estudar): {diferenca if diferenca else '∅'}")
print(f" Diferença Candidato - Vaga (Seus diferenciais): {diferencial if diferencial else '∅'}")
```
