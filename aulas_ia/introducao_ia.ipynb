### Exercicio sobre lógica

Ler nomes e salvar em uma lista quando o nome lido for diferente de sair;

Ordenar na mesma lista (in place);

Mostrar os nomes lidos pelo índice.

-matheus

-caio

-lucas

-sair | SAIR

0 - matheus

1 - caio

2 - lucas

```python
lista_nome = []

while True:
  nome = str(input("Digite o nome ou sair para encerrar: "))
  if nome.lower() == "sair":
    break
  lista_nome.append(nome)
```

```python
lista_nome.sort()

reversed_nome = []
reversed_nome_index = []

for i in range(len(lista_nome) - 1, -1, -1):
  reversed_nome.append(lista_nome[i])

# for i in range(len(lista_nome) - 1, -1, -1):
#   print(f"{i} - {lista_nome[i]}")
#   print(i)
#   reversed_nome_index.insert(i, lista_nome[i])
```

```python
for i in range(len(lista_nome)):
    print(f"{i} - {lista_nome[i]}")

print(f"\n---- Lista invertida ----")
for i in range(len(reversed_nome)):
    print(f"{i} - {reversed_nome[i]}")

# print(f"\n---- Lista invertida com index invertido----")
# for i in range(len(reversed_nome)):
#     print(f"{i} - {reversed_nome_index[i]}")
```
