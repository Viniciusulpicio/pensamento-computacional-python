### Matplotlib


[Matplotlib](https://matplotlib.org/) is a comprehensive library for creating static, animated, and interactive visualizations in Python. Matplotlib makes easy things easy and hard things possible.

```python
import matplotlib.pyplot as plt
```

```python
# Só para garantirmos o estado da variável df
import pandas as pd
df = pd.read_csv('/content/sample_data/california_housing_train.csv')
```

```python
df['median_house_value'].hist()
plt.title('Distribution of Median House Value')
plt.xlabel('Median House Value')
plt.ylabel('Frequency')
plt.show()
```

```python
df.plot.scatter(x='median_income', y='median_house_value')
plt.title('Median House Value vs. Median Income')
plt.xlabel('Median Income')
plt.ylabel('Median House Value')
plt.show()
```

```python
avg_house_value_by_age = df.groupby('housing_median_age')['median_house_value'].mean()
avg_house_value_by_age.plot(kind='bar')
plt.title('Average House Value by Housing Median Age')
plt.xlabel('Housing Median Age')
plt.ylabel('Average House Value')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
scatter = plt.scatter(x=df['longitude'], y=df['latitude'], c=df['median_house_value'], alpha=0.1)
plt.colorbar(scatter, label='Median House Value')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Geospatial Distribution of Housing Prices')
plt.show()
```

```python
correlation_matrix = df.corr()
correlation_matrix
```

```python
plt.figure(figsize=(12, 8))
plt.matshow(correlation_matrix, cmap='coolwarm')
plt.title('Correlation Heatmap of California Housing Dataset')
plt.xticks(range(len(correlation_matrix.columns)), list(correlation_matrix.columns), rotation='vertical')
plt.yticks(range(len(correlation_matrix.columns)), list(correlation_matrix.columns))
plt.colorbar()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
plt.scatter(df['longitude'], df['latitude'],
            c=df['median_house_value'], cmap='jet',
            s=df['population'] / 100, label='Population')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

colorbar = plt.colorbar()
colorbar.set_label('Median House Value', fontsize=12)

plt.legend()
plt.show()
```

### Seaborn

[Seaborn](https://seaborn.pydata.org/) is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Só para garantirmos o estado da variável df
import pandas as pd
df = pd.read_csv('/content/sample_data/california_housing_train.csv')
```

```python
sns.set_style("whitegrid")
```

```python
sns.pairplot(df, diag_kind='kde', corner=True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(df['median_house_value'], bins=50, kde=True)
plt.title("Distribution of Median House Values")
plt.xlabel("Median House Value")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(df["median_income"], bins=50, kde=True)
plt.title("Distribution of Median Income")
plt.xlabel("Median Income")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.scatterplot(x=df['median_income'], y=df['median_house_value'], alpha=0.5)
plt.title("Median Income vs. Median House Value")
plt.xlabel("Median Income")
plt.ylabel("Median House Value")
plt.show()
```

```python
plt.figure(figsize=(8, 8))
sns.scatterplot(x=df["longitude"], y=df["latitude"], hue=df["median_house_value"], palette="coolwarm", alpha=0.5)
plt.title("Geographical Distribution of Houses (Colored by Median House Value)")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.scatterplot(x=df["total_rooms"], y=df["total_bedrooms"], alpha=0.5)
plt.title("Total Rooms vs. Total Bedrooms")
plt.xlabel("Total Rooms")
plt.ylabel("Total Bedrooms")
plt.show()
```

```python
df["income_category"] = pd.qcut(df["median_income"], q=5, labels=["Very Low", "Low", "Medium", "High", "Very High"])

plt.figure(figsize=(10, 5))
sns.boxplot(x=df["income_category"], y=df["median_house_value"])
plt.title("Median House Value by Income Category")
plt.xlabel("Income Category")
plt.ylabel("Median House Value")
plt.show()
```

#### Boxplot

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x=pd.qcut(df['housing_median_age'], q=5), y=df['median_house_value'])
plt.xticks(rotation=45)
plt.title("House Value Distribution by Housing Median Age")
plt.xlabel("Housing Median Age (Binned)")
plt.ylabel("Median House Value")
plt.show()
```

### Extra

Folium builds on the data wrangling strengths of the Python ecosystem and the mapping strengths of the Leaflet.js library. Manipulate your data in Python, then visualize it in a Leaflet map via Folium.

```python
import folium
import numpy as np
from branca.colormap import LinearColormap

map_center = [df['latitude'].mean(), df['longitude'].mean()]
m = folium.Map(location=map_center, zoom_start=6)

colormap = LinearColormap(['blue', 'green', 'yellow', 'red'], vmin=df['median_house_value'].min(), vmax=df['median_house_value'].max())

for i, row in df.iterrows():
    folium.CircleMarker(
        location=[row['latitude'], row['longitude']],
        radius=row['population'] / 1000,
        color=colormap(row['median_house_value']),
        fill=True,
        fill_color=colormap(row['median_house_value']),
        fill_opacity=0.7,
        popup=f"Median House Value: ${row['median_house_value']}, Population: {row['population']}",
    ).add_to(m)

colormap.add_to(m)

m.save("housing_map_with_colors.html")
m
```
