```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("/content/global_movies_dataset_1950_2026.csv")
```

```python
df_limpo = df.dropna(subset=['imdb_rating', 'audience_score']).copy()
df_limpo
```

```python
print(df_limpo.info())
```

```python
X = df_limpo['imdb_rating']
y = df_limpo['audience_score']

plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='purple', alpha=0.3)

plt.title("Relação: Nota do IMDB vs Nota do Público", fontsize=14, fontweight="bold")
plt.xlabel("Nota dos Críticos (IMDB)", fontsize=12)
plt.ylabel("Nota da Audiência (Audience Score)", fontsize=12)
```
