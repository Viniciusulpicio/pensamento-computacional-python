## Exercicio de Inventário

Desenvolva um algoritimo de inventário de RPG. O programa deve exibir um menu com as opções:

  1. Adicionar item: usuário digita o nome de um item (ex: "Pção de Cura") que deve ser guardado em uma lista;
  2. Remover último item: o algoritimo deve remover o último item adicionado
  3. Sair: encerrar

Ao final de cada operação (adicionar ou remover), exiba o estado atual da lista de itens.

```python
itens = []

while True:
  comando = input("1. Adicionar item: usuário digita o nome de um item ex: Poção de Cura que deve ser guardado em uma lista; \n2. Remover último item: o algoritimo deve remover o último item adicionado \n3. Sair: encerrar \nDigite o comando desejado: ")

  if comando == "1":
    adicionar = input("\nDigite o item que quer adicionar a seu inventário: ")
    itens.append(adicionar)
    print(f"Seus itens são: {itens}\n")

  elif comando == "2":
    print("\nUltimo item sendo removido...")
    itens.pop()
    print(f"Seus itens são: {itens}\n")

  elif comando == "3":
    print("\nEncerrando controle de inventario.")
    break

  else:
    print("\DIGITE UM COMANDO VÁLIDO\n")
```
