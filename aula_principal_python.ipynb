## Aprendendo Match Case

```python
numero = int(input("Digite o numero da semana: "))

match numero:
    case 1:
        print("Domingo")
    case 2:
        print("Segunda")
    case 3:
        print("Terça")
    case 4:
        print("Quarta")
    case 5:
        print("Quinta")
    case 6:
        print("Sexta")
    case 7:
        print("Sabado")
```

#### Crie um programa que peça ao usuário para digitar uma opção:

1 -> Financeiro
2 -> Suporte técnico
3 -> Comercial
4 -> Cancelamento

Utilize match case para verificar a opção escolhida e imprimir para o usuário "Você será redirecionado para o setor X"

Caso a opção seja inválida, mostre "Opção inválida, tente novamente"

```python
numero = int(input("1 -> Financeiro \n2 -> Suporte técnico \n3 -> Comercial \n4 -> Cancelamento \nDigite uma opção: "))

match numero:
  case 1:
    print("1 -> Financeiro")
  case 2:
    print("2 -> Suporte técnico")
  case 3:
    print("3 -> Comercial")
  case 4:
    print("4 -> Cancelamento")
  case _:
    print("Opção inválida, tente novamente")
```

```python
n1 = float(input("Digite o primeiro número da operação: "))
n2 = float(input("Digite o segundo número da operação: "))

operacao = str(input("\na -> soma \nb -> subtração \nc -> multiplicação \nd -> divisão \nDigite qual operação quer fazer: "))
operacao = operacao.upper().strip()

match operacao:
  case "A":
    print(f"A soma entre {n1} e {n2} é: {n1+n2:.2f}")
  case "B":
    print(f"A subtração entre {n1} e {n2} é: {n1-n2:.2f}")
  case "C":
    print(f"A multiplicação entre {n1} e {n2} é: {n1*n2:.2f}")
  case "D":
    print(f"A divisão entre {n1} e {n2} é: {n1/n2:.2f}")
  case _:
    print("Operação inválida")
```

#### Crie um programa que:

1. Receba uma nota de 0 a 10 (inteiro)
2. Classifique o aluno usando match case:
- 10 -> "Excelente"
- 0 ou 10 -> "Muito bom"
- 6 ou 7 -> "Aprovado"
- 0 a 5 -> "Reprovado"

Dica: para usar o "ou" no match case, use |


```python
nota = int(input("Digite a nota: "))

match nota:
  case 10:
    print("Excelente")
  case 8 | 9:
    print("Muito bom")
  case 6 | 7:
    print("Aprovado")
  case 0 | 1 | 2 | 3 | 4 | 5:
    print("Reprovado")
```

### Exercicio

Faça um algoritimo que peça para o usuário digitar um divisor (número inteiro). A função deve percorrer a lista [10,'20',0,50,'texto',30]. Para cada item da lista, tente dividir o item pelo divisor que o usuário informou. Você deve tratar especificamente os seguintes erros:

- Erro de tipo
- Erro de Valor/Conversão
- Erro de divisão por zero
- Erro genérico

```python
lista = [10, '20', 0, 50, 'texto', 30]

try:
    divisor = int(input("Digite um número inteiro: "))

    for item in lista:
        try:
            divisao = item / divisor
            print(f"Resultado: {divisao}")

        except ZeroDivisionError:
            print(f"Erro: Não é possível dividir por zero.")
        except ValueError:
            print(f"Erro: O valor '{item}' não pode ser convertido para número.")
        except TypeError:
            print(f"Erro de tipo ao processar o valor.")
        except Exception as e:
            print(f"Ocorreu um erro inesperado: {e}")

except ValueError:
    print("Erro crítico: Você precisa digitar um número inteiro válido para o divisor.")
```

## Operadores de conjuntos


| Operação | Operador | Descrição | Exemplo (`a={1,2}`, `b={2,3}`) |
| :--- | :---: | :--- | :--- |
| **União** | `\|` | Retorna todos os elementos de ambos os conjuntos. | `a \| b` -> `{1, 2, 3}` |
| **Interseção** | `&` | Retorna apenas os elementos que existem em ambos. | `a & b` -> `{2}` |
| **Diferença** | `-` | Retorna os elementos do primeiro que não estão no segundo. | `a - b` -> `{1}` |
| **Diferença Simétrica** | `^` | Retorna elementos que estão em um ou outro, mas não nos dois. | `a ^ b` -> `{1, 3}` |
| **Subconjunto** | `<=` | Verifica se o conjunto da esquerda está contido no da direita. | `a <= b` -> `False` |
| **Superconjunto** | `>=` | Verifica se o conjunto da esquerda contém o da direita. | `a >= {1}` -> `True` |

```python
compras = ["arroz", "feijao", "macarrão", "leite", "arroz", "leite"]

conj_compras = set(compras)
print(conj_compras)
```

```python
compras = ["arroz", "feijao", "macarrão", "leite", "arroz", "leite"]

for c in compras:
  conj_compras.add(c)

print(conj_compras)
```

```python
lista1 = ["titanic" "as branquelas", "os vingadores", "batman", "a lagoa azul"]
lista2 = ["as branquelas", "batman", "o diabo veste prada"]

set_lista1 = set(lista1)
set_lista2 = set(lista2)

todos_produtos = set_lista1 | set_lista2
em_comum = set_lista1 & set_lista2
apenas_lista1 = set_lista1 - set_lista2
diferenca_simetrica = set_lista1 ^ set_lista2

print(f"Todos os filmes: {todos_produtos}")
print(f"Filmes em comum: {em_comum}")
print(f"Filmes apenas da lista 1: {apenas_lista1}")
print(f"Diferença simetrica: {diferenca_simetrica}")
```
