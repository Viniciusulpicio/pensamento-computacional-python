## Input

Input() é a função que recebe uma informação vinda do usuario isso permite a atribuição de variaveis de maneira dinamica

```python
# APRENDENDO INPUT

nome = input("Qual o seu nome? ")
print("Bem vindo ", nome)
```

```python
# DECOMPONDO COM INPUT

numero = int(input("Digite um valor para decompor: "))

centena = numero  // 100
dezena = (numero % 100) // 10
unidade = (numero % 10)

print ("Possui", centena, "centenas")
print ("Possui", dezena, "dezenas")
print ("Possui", unidade, "unidades")
```

```python
#1  Receba por input dois numeros e mostre: soma, subtração, divisão e multiplicação

print("soma, subtração, divisao e multiplicacao de valores")
n1 = int(input("Digite o primeiro valor: "))
n2 = int(input("Digite o segundo valor: "))

sum = n1+n2
sub = n1-n2
div = n1/n2
mult = n1*n2

print("Soma: ", sum)
print("Subtração: ", sub)
print("Divisão: ", div)
print("Multiplicação", mult)

print("\n" )
# 2 converta as unidades de (minutos -> segundo )

print("conversor de minutos para segundos")
min = int(input("Digite os minutos: "))

convert = min*60
print("Em", min, "minutos existem", convert, "segundos")

print("\n" )
# 3 faça uma média de 3 notas

print("media de 3 notas")
n1 = int(input("Digite a primeira nota: "))
n2 = int(input("Digite a segunda nota: "))
n3 = int(input("Digite a terceira nota: "))

media = (n1+n2+n3)/3
print("A média das 3 notas é: ", media)

print("\n" )
# 4 receba o preço de um produto e calcule o valor do troco

print("calcular valor do troco")
prod = int(input("Qual valor do produto: "))
receb = int(input("Qual valor recebido: "))

troco = (receb-prod)
print("Deve ser dado de troco", troco, "reais")

print("\n" )
# 5 calcule o total de em segundos ( receba hora, minuto e segundo e calcule quantos segundos existem no total)]

print("calcular hora, minuto e segundo")
hora = int(input("Digite a hora: "))
min = int(input("Digite os minutos: "))
segundos = int(input("Digite os segundos: "))

mult = (hora*(60*60))+(min*60)+segundos

print(mult)
```
