```python
import pandas as pd
```

```python
# exclui tudo que tem 1 dado nulo na linha
df = df.dropna()

# tipos de dados
# df["Data"] = pd.to_datetime(df["Data"], format="%Y-%m-%d")

# padronizacao
df["Occupation"] = df["Occupation"].str.strip()
df["Occupation"] = df["Occupation"].str.title()

# duplicatas
df = df.drop_duplicates(subset=["Person_ID"])

display(df)
display(df.info())
```
