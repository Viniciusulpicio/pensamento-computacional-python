```python
print(f"Olá, sou seu assistente pessoal Jarvis, vamos organizar sua rotina hoje !")
nome_user = str(input("Qual seu nome? "))
print(f"Seja bem-vindo {nome_user}\n")

tarefas = []

while True:
  print("------------------------------")
  print(f"      MENU PRINCIPAL {nome_user.upper()}")
  print("------------------------------")
  acao = str(input("1. Ver tarefas \n2. Adicionar tarefa \n3. Remover tarefa \n4. Limpar toda a lista \n5. Sair \nO que vamos fazer hoje: "))
  acao = acao.upper().strip()

  if acao == "1":
    print(f"\n--- SUAS TAREFAS {nome_user.upper()} ---")
    if not tarefas:
            print("Sua lista está vazia!")
    else:
      for i, tarefa in enumerate(tarefas):
        print(f"Posição {i+1}: {tarefa}")
    print("------------------------------\n")

  elif acao == "2":
    elemento_add = str(input("\nQual tarefa quer adicionar na lista: "))
    if elemento_add:
      tarefas.append(elemento_add)
      print(f"Tarefa adicionada {elemento_add}")
    else:
      print("A tarefa está vazia")
    for i, tarefa in enumerate(tarefas):
      print(f"Posição {i+1}: {tarefa}")
    print("------------------------------\n")

  elif acao == "3":
    if not tarefas:
            print("Sua lista está vazia!")
    else:
      for i, tarefa in enumerate(tarefas):
        print(f"Posição {i+1}: {tarefa}")

    elemento_remove = int(input("\nEscreva o numero da tarefa que quer eliminar: "))
    del tarefas[elemento_remove-1]

    for i, tarefa in enumerate(tarefas):
      print(f"Posição {i+1}: {tarefa}")

  elif acao == "4":
    confirmar = str(input("Tem certeza que deseja apagar todas suas tarefas (S/N): ")).upper().strip()
    if confirmar == "S":
      tarefas.clear()
      print("Tarefas vazias")

  elif acao == "5":
    print(f"\nAté logo, {nome_user}! Jarvis encerrando...")
    break

  else:
    print("\nComando errado, tente novamente\n")
```
