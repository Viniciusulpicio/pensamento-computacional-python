#### Contexto: Um banco financeiro precisa de um algoritmo que receba dados de um

cliente e determine qual o melhor produto financeiro para ele.

Requisitos do Programa:

1. Entrada de Dados:

○ Nome (String)

○ Idade (Inteiro)

○ Salário Mensal (Real)

○ Possui Nome Limpo? (Booleano - use True/False ou 1/0)

○ Valor que deseja investir (Real)

#### 2. Processamento:

○ Cálculo de Score: O sistema deve calcular se o cliente é "Elegível para
Crédito".

■ Regra 1: Ser maior de 18 anos

■ Regra 2: Possuir nome limpo

○ Classificação de Investimento: O sistema deve calcular o nível do
investidor de acordo com a porcentagem do seu salário que ele quer
investir

■ Investidor Iniciante: valor de investimento for menor que 10% do
salário seu mensal

■ Investidor Moderado: valor de investimento for entre 10% e 30%
do salário seu salário mensal

■ Investidor Arrojado: valor de investimento for acima de 30% do
salário seu salário mensal
○ Bônus Especial: Se o cliente for elegível para crédito e for um
investidor arrojado, ele ganha o selo "Cliente VIP".

####3. Saída de Dados:

○ Um resumo formatado contendo o Nome, a Classificação de Investidor
e se ele é ou não um "Cliente VIP".

```python
nome = str(input("Digite seu nome: "))
idade = int(input("Digite sua idade: "))
salario = float(input("Digite seu salario em real: "))
nome_limpo = str(input("Digite sim ou não para informar se seu nome é limpo: ")).lower().strip()
investimento = float(input("Digite o valor que quer investir: "))

if idade < 18 or nome_limpo != "sim":
  print("Usuario não é elegível para crédito")
else:
  if investimento >= salario*0.30:
    nivel_investidor = "Investidor Arrojado"
    cliente_vip = "Cliente VIP"
  elif investimento < salario*0.10:
    nivel_investidor = "Investidor Iniciante"
    cliente_vip = "Não é um cliente VIP"
  elif investimento >= salario*0.10 or investimento < salario*0.30:
    nivel_investidor = "Investidor Moderado"
    cliente_vip = "Não é um cliente VIP"

  print("Olá ", nome, "sua classificação de investimento é:", nivel_investidor, "seu nivel de cliente é:", cliente_vip)
```
